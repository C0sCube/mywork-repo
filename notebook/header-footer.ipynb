```python
import pdfplumber

def extract_text_without_header_footer(path):
    with pdfplumber.open(path) as pdf:

        header_bottom = None
        footer_top = None
        
        # List filtered text
        filtered_text = []

        pages = list(pdf.pages)

        #print(pages[0])
        
       #Analyze the first 5 pages to obtain header footer data
        for i, page in enumerate(pages[:5]):  
            words = page.extract_words()
            
            # print(f"Page {i+1} - First few words:")
            # for word in words[:5]:  # Print first 5 words
            #     print(word)
            
            # print(f"Page {i+1} - Last few words:")
            # for word in words[-5:]:  # Print last 5 words
            #     print(word)

            # Set header and footer positions based on the first page
            if i == 0:
                header_bottom = words[0]["bottom"]
                footer_top = words[-1]["top"]

                # print(f"Header ends at: {header_bottom}")
                # print(f"Footer starts at: {footer_top}")

        # Extract text from all pages while omitting the header and footer
        for page in pages:
            words = page.extract_words()
            page_text = []

            for word in words:
                if word["top"] > header_bottom and word["bottom"] < footer_top:
                    page_text.append(word)
            filtered_text.append(page_text)

    return filtered_text
```
