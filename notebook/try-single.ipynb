```python
def extract_pdf(input_pdf_path, pages_to_extract):
    input_doc = fitz.open(input_pdf_path)
    extracted_data = []

    with pdfplumber.open(input_pdf_path) as pdf:
        total_pages = len(pdf.pages)  # Total pages

        for page_number in range(total_pages):
            if any(isinstance(spec, tuple) and spec[0] <= page_number <= spec[1] for spec in pages_to_extract if isinstance(spec, tuple)) \
                    or page_number in pages_to_extract:
                doc_page = input_doc[page_number]
                blocks = doc_page.get_text("dict")["blocks"]
                sorted_blocks = sorted(blocks, key = lambda x: (x['bbox'][1], x['bbox'][0])) #sort T to B , L to R
                text = doc_page.get_text("text")

                # Store each block along with the page number
                page_data = {
                    "page_number": page_number,
                    "blocks": sorted_blocks,
                    "text": text
                }
                extracted_data.append(page_data)

    input_doc.close()
    return extracted_data
```

```python
import fitz  # Import PyMuPDF

def extract_fonts_from_text_blocks(pdf_path):
    doc = fitz.open(pdf_path)  # Open the PDF file
    fonts_used = set()  # Use a set to avoid duplicate entries

    for page_number in range(len(doc)):
        page = doc[page_number]
        text_dict = page.get_text("dict")  # Get text content as a dictionary
        blocks = text_dict['blocks']  # Retrieve blocks from the text dictionary

        for block in blocks:
            if 'lines' in block:  # Ensure the block contains lines
                for line in block['lines']:
                    for span in line['spans']:  # Iterate through each span in the line
                        font_info = (span['font'], round(span['size']), span['flags'])
                        fonts_used.add(font_info)  # Store font name, size, and style flags

    doc.close()
    return fonts_used

# Example usage
pdf_path = path + samco_path # Ensure this path is correct
font_details = extract_fonts_from_text_blocks(pdf_path)
for font in font_details:
    print(f"Font: {font[0]}, Size: {font[1]}, Flags: {font[2]}")
```
