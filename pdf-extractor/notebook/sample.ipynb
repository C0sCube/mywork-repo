```python
import fitz  # PyMuPDF

# Open a document
doc = fitz.open("your_document.pdf")

# Get the blocks of the first page
page = doc.load_page(0)
blocks = page.get_text("dict")["blocks"]

# Iterate over blocks
for block in blocks:
    print(f"Block type: {block['type']}")
    print(f"Bounding box: {block['bbox']}")
    
    if block['type'] == 0:  # Text block
        print("Text in block:", block['text'])
        for line in block['lines']:
            print("Line bbox:", line['bbox'])
            print("Line text:", " ".join(span['text'] for span in line['spans']))
            for span in line['spans']:
                print("Span text:", span['text'])
                print("Font:", span['font'])
                print("Font size:", span['size'])
    elif block['type'] == 1:  # Image block
        print(f"Image size: {block['width']}x{block['height']}")
        print(f"Image bounding box: {block['bbox']}")
```

```python
pdf = fitz.open(path)
total_page = len(pdf)
print(total_page)

third_block = list()

for pg in range(1, 4):
    #traverse each page to get blocks
    #pdfr->page
    page = pdf[pg]
    blocks = page.get_text('dict')['blocks']
    sorted_blocks = sorted(blocks, key = lambda b: (b['bbox'][1], b['bbox'][0]))
    tot_block_per_page = len(blocks)
    print(tot_block_per_page,"\n")

    if pg == 3:
        third_block = sorted_blocks


    #pdf->page->block
    for i,block in enumerate(sorted_blocks):
        print(f"\n block{1+i}:{block['type']}")
```
